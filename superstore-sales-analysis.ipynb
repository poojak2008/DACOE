```python
#Extracted new column as "Year" from "Order Date"

df["Year"] = df["Order Date"].astype(str).str[6:10]
df
```

```python
#Unique category of products

df['Category'].unique()
```

```python
#Type of varities in sub category column

df['Sub-Category'].unique()
```

```python
#Getting the sales towards each sub category which belongs to a specific category

df.groupby(['Category','Sub-Category'])['Sales'].sum().reset_index()
```

**1. Show the distribution of segment by pie plot.**

```python
# Number of customers in each segment 

segment_type = df['Segment'].value_counts().reset_index() 
segment_type = segment_type.rename(columns={'Segment': 'Customer Type', 'count': 'Total Customers'})

segment_type
```

```python
plt.pie(segment_type['Total Customers'], labels = segment_type['Customer Type'], autopct='%1.1f%%')

plt.title('Distribution of categories')
plt.show()
```

**2. Show the sales for each category.**

```python
segment_sales = df.groupby('Segment')['Sales'].sum().reset_index()
segment_sales = segment_sales.rename(columns={'Segment': 'Customer Type', 'Sales': 'Total Sales'})
segment_sales

plt.bar(segment_sales['Customer Type'], segment_sales['Total Sales'])
plt.title('Sales data for each segment')
plt.xlabel('Segment Type')
plt.ylabel('Total Sales')
plt.show()
```

**3. Which category of shipping contains the highest percentage ratio?**

```python
shipping_mode = df['Ship Mode'].value_counts().reset_index()
shipping_mode = shipping_mode.rename(columns={'Ship Mode':'Mode_of_Shipment', 'count': 'Use Frequency'})

plt.pie(shipping_mode['Use Frequency'], labels=shipping_mode['Mode_of_Shipment'], autopct='%1.1f%%')
plt.title('Contribution of shipping mode in %')
plt.show()
```

**4. Show the sales for each category by bar plot?**

```python
highest_sale_subcategory = df.groupby(['Sub-Category'])['Sales'].sum().reset_index()
highest_sale_subcategory = highest_sale_subcategory.sort_values(by='Sales', ascending=False) 
highest_sale_subcategory.reset_index(drop=True)
highest_sale_subcategory = highest_sale_subcategory.sort_values(by='Sales', ascending=True)
plt.barh(highest_sale_subcategory['Sub-Category'], highest_sale_subcategory['Sales'])
plt.title('Category wise sales')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()
```

**5. Create a line chart for sales over year.**

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
year_wise_sale = df.groupby(df['Order Date'].dt.year)['Sales'].sum()
year_wise_sale = year_wise_sale.reset_index()
year_wise_sale = year_wise_sale.rename(columns={'Order Date': 'Year', 'Sales':'Total Sales'})
year_wise_sale
plt.plot(year_wise_sale['Year'], year_wise_sale['Total Sales'], marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total sales by year')
plt.tight_layout()
plt.show()
```
